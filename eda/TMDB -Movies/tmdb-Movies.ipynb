Data cleaning process by filling null values with mean and also drop duplicate data

```python
#drop  unuseful columns
df.drop(['id','imdb_id','homepage','overview'],axis=1, inplace= True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Drop nul values for each columns containing null values
df.dropna(inplace= True)
df.info()
```

```python
df.shape
```

```python
#To check null values
df.describe()
```

```python
#to replace all numeric column which has zero value in column with the mean value
df['popularity']= df['popularity'].replace(0,df['popularity'].mean())
df['budget'] =df['budget'].replace(0,df['budget'].mean())
df['revenue'] =df['revenue'].replace(0,df['revenue'].mean())
df['runtime'] =df['runtime'].replace(0,df['runtime'].mean())
df['budget_adj'] =df['budget_adj'].replace(0,df['budget_adj'].mean())
df['revenue_adj'] =df['revenue_adj'].replace(0,df['revenue_adj'].mean())
```

```python
#replace  all Zeros values with mean value 
df.describe()
```

```python
#Calculate sum of all duplicated value
df.duplicated().sum()
```

```python
#Drop duplicate vaue
df.drop_duplicates(inplace=True)
```

# Evaluating question 

1.Does the higher budget mean higher popularity?


```python
#plot the relation between budgetand popularity

x=df['budget']
y= df['popularity']
plt.scatter(x,y)
plt.title("Average Popularity by Difference Budget")
plt.xlabel('budget', fontsize= 12)
plt.ylabel('popularity',fontsize= 12)
```

As per the above scatter plot its very difficult to observe strong relationship between popularity ans
budget. so we use other method to observe the relationship between them. In
this method we divide dataset  into  two group on the basis of median.

```python
#Divide  the budget into two groups : lesser_cost and more_cost
med= df['budget'].mean()
lesser_cost= df.query('budget <{}'.format(med))
more_cost= df.query('budget >= {}'.format(med))
```

```python
lesser_cost
```

```python
more_cost
```

```python
#check Lesser cost and more cost mean values for popularity
mean_high_budget= more_cost['popularity'].mean()
mean_low_budget= lesser_cost['popularity'].mean()
```

```python
mean_high_budget
```

```python
mean_low_budget
```

```python
#Create a bar chart with the values we get above
locations= [1,2]
heights= [mean_low_budget, mean_high_budget]
labels=['low','high']
plt.bar(locations,heights, tick_label= labels)
plt.title("Mean popularity by all Budget")
plt.xlabel('Budgets')
plt.ylabel('Mean Popularity')
```

```python
increase_percantage= (mean_high_budget- mean_low_budget)/ mean_high_budget * 100
increase_percantage
```

Answer of the question 1

From the above bar plot we conclude that higher budget movie gains higher popularity .Higher budget movie have Mean popularity more than the Mean popularity of lower budget movie.

# EVALUATING QUESTON 3

3. Is Higher popularity means higher profits?

```python
#Calculation for the mean of popularity
mean= df['popularity'].median()
lowest_popularity= df.query('popularity < {}'.format(mean))
more_popularity= df.query('popularity >= {}'. format(mean))
```

```python
lowest_popularity['popularity'].sum()
more_popularity['popularity'].sum()
```

```python
#Create a new column for profit
df['profit']= df['revenue']- df['budget']
df.head()
#df['profit'].sum()
```

```python
#Average net profit for low_popularity and high_popularity
mean_profit_of_low=  lowest_popularity['profit'].mean()
mean_profit_of_high = more_popularity['profit'].mean()
df.head()
```

```python
print(mean_profit_of_low)
print(mean_profit_of_high)
```

```python
#Create a bar chart with the values we get above 

locations= [1,2]
heights= [mean_profit_of_low,mean_profit_of_high]
labels= ['Low','High']
plt.bar(locations,heights, tick_label=labels)
plt.title('Average profit by Different Popularity', fontsize= 15)
plt.xlabel('Popularity')
plt.ylabel('Average Profit')
```

ANSWER FOR QUESTION 3

From the above graph we observe that higher popularity leads to more Average profit.
